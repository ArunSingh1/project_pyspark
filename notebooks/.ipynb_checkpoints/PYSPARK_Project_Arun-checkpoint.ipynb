```python
from twython import TwythonStreamer
import csv
from unidecode import unidecode
from dateutil import parser

    
def process_tweet(tweet):
    
    #"process the input json data"
        
    dict_ = {'tweet_id': [], 'user': [], 'user_id' : [] , 'created_at_str' : [], 'date': [], 'text': [], 'hashtags' : [], 'lang' : [], 'profilepic': [], 
         'location': [], 'place': [], 'retweet_count': [], 'is_Verfied' : [], 'followers_count':[],  'joined_in' : [], 'geo': [], 'source': []}
    
    dict_['tweet_id']=(tweet['id'])
    dict_['user']=(tweet['user']['screen_name'])
    dict_['user_id']=(tweet['user']['id'])
    dict_['created_at_str']=(tweet['created_at'])
    dict_['date']=parser.parse(tweet['created_at'])
    dict_['text']=(tweet['text']) #unidecode
    dict_['hashtags']=[hashtag['text'] for hashtag in tweet['entities']['hashtags']]
    dict_['lang']=(tweet['user']['lang'])
    dict_['profilepic']=(tweet['user']['profile_image_url'])
    dict_['location']=(tweet['user']['location'])
    dict_['place'] = (tweet['place'])
    dict_['retweet_count']=(tweet['retweet_count'])
    dict_['is_Verfied']=(tweet['user']['verified'])
    dict_['followers_count']=(tweet['user']['followers_count'])
    dict_['joined_in']=(tweet['user']['created_at'])
    dict_['geo']=(tweet['coordinates'])
    dict_['source']=(tweet['source'])

    return dict_
    
```

```python
# Create a class that inherits TwythonStreamer
class MyStreamer(TwythonStreamer):     

    # Received data
    def on_success(self, data):

        # Only collect tweets in English
        if 'text' in data:
            tweet_data = process_tweet(data)
            #print(tweet_data)
            #df1 = pd.DataFrame(tweet_data)
            self.save_to_aws(tweet_data)

    # Problem with the API
    def on_error(self, status_code, data):
        print(status_code, data)
        self.disconnect()
        
    # Save each tweet to csv file in AWS S3 bucket
    def save_to_aws(self, tweet):
        with open(r'geo1.csv', 'a') as file:
            writer = csv.writer(file)
            writer.writerow(list(tweet.values()))
            df = pd.read_csv('geo1.csv', header=None)
            
            
            bucket = 'mybucket-147'# already created on S3
            csv_buffer = StringIO()
            df.to_csv(csv_buffer)
            
            ACCESS_KEY = os.environ.get('AWSAccessKeyId')
            SECRET_KEY = os.environ.get('AWSSecretKey')
            
            s3 = boto3.client('s3', aws_access_key_id=ACCESS_KEY,
                      aws_secret_access_key=SECRET_KEY)
            
            
            s3_resource = boto3.resource('s3',aws_access_key_id=ACCESS_KEY,
                      aws_secret_access_key=SECRET_KEY)
        
            s3_resource.Object(bucket, 'rawtweets.csv').put(Body=csv_buffer.getvalue())

#     def to_s3_bucket(self, tweet):
#         bucket = 'mybucket-147'# already created on S3
#         csv_buffer = StringIO()
#         df.to_csv(csv_buffer)
#         s3_resource = boto3.resource('s3')
#         s3_resource.Object(bucket, 'geo1.csv').put(Body=csv_buffer.getvalue())
        
```

### Processing data from Sentiment Analysis

```python
tw_list = sqlc.sql("select text from tweet").toPandas()
```

```python
tw_list
```

```python
import string

def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    #text = re.sub('@[a-z]*[0-9]+', '', text)
    text = re.sub('[^A-Za-z0-9]+',  ' ', text)
    text = text.replace("RT", "")
    #text = re.sub('[0-9]+', '', text)
    return text

tw_list['text_cleaned'] = tw_list['text'].apply(lambda x: remove_punct(x))
tw_list.head(10)

```

```python
# !pip install matplotlib
```

```python
# Import Libraries
from textblob import TextBlob
import sys
import matplotlib.pyplot as plt
import nltk
import re
import string
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.stem import SnowballStemmer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
tw_list[['polarity', 'subjectivity']] = tw_list['text'].apply(lambda Text: pd.Series(TextBlob(Text).sentiment))
```

```python
for index, row in tw_list['text'].iteritems():
     score = SentimentIntensityAnalyzer().polarity_scores(row)
     neg = score['neg']
     neu = score['neu']
     pos = score['pos']
     comp = score['compound']
     if neg > pos:
         tw_list.loc[index, 'sentiment'] = 'negative'
     elif pos > neg:
         tw_list.loc[index, 'sentiment'] = 'positive'
     else:
         tw_list.loc[index, 'sentiment'] = 'neutral'
         tw_list.loc[index, 'neg'] = neg
         tw_list.loc[index, 'neu'] = neu
         tw_list.loc[index, 'pos'] = pos
         tw_list.loc[index, 'compound'] = comp
tw_list.head(10)
```

### Upload the Processed data to AWS S3 bucket 

```python
import boto3
from botocore.exceptions import NoCredentialsError

#Credentials AWS


# UPLOAD TO AWS  
def upload_to_aws(local_file, bucket, s3_file):
    s3 = boto3.client('s3', aws_access_key_id=ACCESS_KEY,
                      aws_secret_access_key=SECRET_KEY)

    try:
        s3.upload_file(local_file, bucket, s3_file)
        print("Upload Successful")
        return True
    except FileNotFoundError:
        print("The file was not found")
        return False
    except NoCredentialsError:
        print("Credentials not available")
        return False
    
uploaded = upload_to_aws('/home/arun/Documents/twitter/notebooks/processed_tweets.csv', 'mybucket-147', 'processed_tweets.csv')
```
